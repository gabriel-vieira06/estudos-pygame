```python
import pygame
from pygame.locals import *
from sys import exit

pygame.init()

largura = 640
altura = 480

x = largura/2 - 20
y = altura/2 - 25

tela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption("Titulo da Janela")
relogio = pygame.time.Clock()

while True:
    relogio.tick(30)
    tela.fill((0,0,0))
    
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            exit()
            
        """
        Trata para eventos onde uma tecla foi pressionada:
        - Tecla a: o objeto é deslocado para esquerda
        - Tecla s: o objeto é deslocado para baixo
        - Tecla d: o objeto é deslocado para direita
        - Tecla w: o objeto é deslocado para cima
        
        O número de pixels deslocados é arbitrário
        (nesse caso são 20)
        
        Problema 1: Manter a tecla pressionada, não faz com que
        o objeto continue se movendo
        
        Problema 2: O objeto pode sair dos limites da janela
        
        if event.type == KEYDOWN:
            if event.key == K_a:
                x = x - 20
            if event.key == K_s:
                y = y + 20
            if event.key == K_d:
                x = x + 20
            if event.key == K_w:
                y = y - 20
        """
    
    """
    A solução abaixo também trata para teclas pressionadas, mas corrige
    ambos os problemas da solução anterior
    """
    
    if pygame.key.get_pressed()[K_a]:
        x = x - 20
        if x <= 0:
            x = largura
            
    if pygame.key.get_pressed()[K_s]:
        y = y + 20
        if y >= altura:
            y = 0
            
    if pygame.key.get_pressed()[K_d]:
        x = x + 20
        if x >= largura:
            x = 0
        
    if pygame.key.get_pressed()[K_w]:
        y = y - 20
        if y <= 0:
            y = altura
    
    pygame.draw.rect(tela, (255,0,0), (x,y,40,50))

    pygame.display.update()
```
